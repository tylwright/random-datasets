```javascript
const fs = require('fs');

let data = fs.readFileSync('Starship Classes.json');

console.log(JSON.parse(data));
```

```javascript
console.log(JSON.parse(data)['operator']['Federation']);
```

```javascript
federationStarships = JSON.parse(data)['operator']['Federation'];

for (starship in federationStarships) {
    console.log(federationStarships[starship]['name']);
}
```

```javascript
federationStarships = JSON.parse(data)['operator']['Federation'];

for (starship in federationStarships) {
    if (federationStarships[starship]['type']) {
        console.log(federationStarships[starship]['name'] + " (" + federationStarships[starship]['type'] + ")");
    } else {
        console.log(federationStarships[starship]['name']);
    }
}
```
